```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cot_df = pd.read_csv('indexData.csv')

display(cot_df)
```

```python
cot_df = cot_df[['Index', 'Date', 'Close']]

display(cot_df)
```

```python
remove = cot_df.loc[cot_df['Index'] != 'NYA']

cot_df = cot_df.drop(remove.index)

cot_df = cot_df[1:11]

display(cot_df)
```

```python
figura = plt.figure(figsize=[20, 3])

plt.plot(cot_df['Date'], cot_df['Close'], '-.', c='#36EC97', label='Date/Close')
plt.xlabel('Date')
plt.ylabel('Close')
figura.suptitle('Bolsa de valores NYA')

plt.legend(fontsize=15)

plt.show(figura)
```
